Question 03:

```python
dict_data_03 = {
    "day" : [f"1/{i}/2017" for i in range(7) if i != 0],
    "temperature" : [32, 35, 28, 24, 32, 31],
    "windspeed" : [6, 7, 2, 7, 4, 2],
    "event" : ["Rain", "Sunny", "Snow", "Snow", "Rain", "Sunny"],
}
solution03 = pd.DataFrame(data=dict_data_03)
solution03
```

Question 04:

```python
new_index = pd.Series(['a','b','c','d','e','f'])
solution04 = solution03.set_index(new_index) #inplace=true will make changes to the solution03 dataframe
solution04
```

Question 05:

```python
print("Mean:", solution03["temperature"].mean())
print("Maximum:", solution03["temperature"].max())
print("Minimum:", solution03["temperature"].min())
```

```python
columns_to_import = ["First Name", "Sex", "Email", "Phone", "Job Title"]
df_people = pd.read_csv("people.csv",
                        index_col=["Sex", "Job Title"],
                        usecols=columns_to_import,
                        skiprows=[1,5])
df_people
```
