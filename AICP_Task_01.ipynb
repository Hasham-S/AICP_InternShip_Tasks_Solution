**Question 03: How to compute the cross-product of two matrices in NumPy?**

```python
# Since the vector/cross product is reserved only for vectors
# The vector product of matrices isn't possible through conventional methods
# however numpy has a different take on this
# it calculate the cross-product of the corresponding row vectors in both the matrices

matrix_A = np.array([
    [1, 2, 3],
    [1, 2, 3]
])
matrix_B = np.array([
    [4, 5, 6],
    [4, 5, 6]
])

cross_product = np.cross(matrix_A, matrix_B)
cross_product
```

**Question 04: How to compute the determinant of an array using NumPy?**

```python
# the expected answer is 63
A = np.array([
    [1, 3, -2],
    [0, 5, 1],
    [4, -1, 2]
])
det_A = np.linalg.det(A)
det_A
```

**Question 06: How to create a 5x5 array with random values and find the minimum and maximum values using
NumPy?**


```python
random_array = np.random.random(size=(5,5))
print(random_array)

print("Max:", random_array.max())
print("Min:", random_array.min())
```

**Question 07: How to compute the mean, standard deviation, and variance of a given array along the second axis in
NumPy?**

```python
np.random.seed(10) # the seed is for generating the same random array every time
random_array = np.random.randint(1, 10, size=(5,5))
print(random_array)

print("Mean: ", random_array.mean(axis=1)) # here (axis=1) means 2nd axis
print("Standard Deviation: ", random_array.std(axis=1))
print("Variance:", random_array.var(axis=1))
```
