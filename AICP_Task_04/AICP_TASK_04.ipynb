Q.7: Draw a horizontal bar graph for items in each category against the calories.
Here is one sample for all items in Beef & Pork, you have to do it for items in each category. Also, write
your observation.

```python
sorted_data = data.sort_values(by=["Calories"], ascending=False) # sorting the data
category = sorted_data["Category"].drop_duplicates() # removing duplicates

for column in category.values:
  # if sorted_data["category"] is same as that of
  # column than select item and calories row of that datapoint
  Y = sorted_data[sorted_data["Category"] == column]["Item"]
  X = sorted_data[sorted_data["Category"] == column]["Calories"]
  fig = px.bar(
      x=X,  # plotting x
      y=Y,  # plotting y
      orientation='h', # horizontal orientation
      title=column
  )
  fig.update_layout( # update layout for renaming the x and y axis
    xaxis_title="Calories",
    yaxis_title=column
  )
  fig.show()
```
